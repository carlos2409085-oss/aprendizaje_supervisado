```python
plt.figure(figsize = (12, 8))

# colores para cada una de las clases
color = ['#240750', '#344C64']

sns.countplot(df, x = 'Embarked', hue = 'Survived', palette=color) 
plt.legend(title = 'Supervivencia', labels = ['No sobrevivi', 'S sobrevivi'])
plt.show()
```

```python
plt.figure(figsize = (12, 8))

sns.histplot(df, x = 'Age', kde = True)
plt.title('Distribucin de Edades')
plt.axvline(df['Age'].mean(), color = 'red')
plt.axvline(df['Age'].median(), color = 'green')
plt.axvline(df['Age'].mode()[0], color = 'black')
plt.show()

# Print the results
print(f"Media: {df['Age'].mean()}")
print(f"Mediana: {df['Age'].median()}")
print(f"Moda: {df['Age'].mode()[0]}")
```

```python
plt.figure(figsize = (12, 8))
sns.heatmap(df[['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']].corr(), 
            annot=True, 
            cbar=True,  # Activar la barra de colores
            cmap='coolwarm',  # Cambiar la paleta de colores
            cbar_kws={'label': 'Nivel de correlacin'})  # Etiqueta para la barra de colores
plt.title('Correlacin de variables numricas')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 1. Grfica de distribucin de edades
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], kde=True, color='skyblue')
plt.title('Distribucin de Edades de los Pasajeros')
plt.show()
```

```python
# 2. Cantidad de sobrevivientes por clase (Pclass)
plt.figure(figsize=(8, 5))
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title('Sobrevivientes por Clase (0=No, 1=S)')
plt.show()
```

```python
#3. Cantidad de supervivientes por genero
plt.figure(figsize=(8, 5))
sns.barplot(x='Sex', y='Survived', data=df, estimator=sum, errorbar=None)
plt.title('Cantidad Total de Sobrevivientes por Gnero')
plt.ylabel('Cantidad de Sobrevivientes')
plt.show()
```

```python
# 4. Cantidad de sobrevivientes por puerto (Embarked)
plt.figure(figsize=(8, 5))
sns.barplot(x='Embarked', y='Survived', data=df, estimator=sum, errorbar=None)
plt.title('Cantidad de Sobrevivientes por Puerto de Embarque')
plt.show()
```

```python
# 5. Tasa de supervivencia por grupo de edad
bins = [0, 1, 12, 18, 30, 60, 120]
labels = ['Infante', 'Nio', 'Adolescente', 'Joven Adulto', 'Adulto', 'Mayor']
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

plt.figure(figsize=(10, 6))
sns.barplot(x='AgeGroup', y='Survived', data=df, errorbar=None, palette='viridis', legend=False)
plt.title('Tasa de Supervivencia por Grupo de Edad')
plt.ylabel('Tasa de Supervivencia (0 a 1)')
plt.show()
```

```python
# 6. Conteo de pasajeros en cada clase (Pclass)
plt.figure(figsize=(8, 5))
sns.countplot(x='Pclass', data=df, palette='Set2')
plt.title('Conteo de Pasajeros por Clase')
plt.show()

```

```python
# 7. Distribucin de gnero dentro de cada clase (Pclass)
plt.figure(figsize=(8, 5))
sns.countplot(x='Pclass', hue='Sex', data=df)
plt.title('Distribucin de Gnero por Clase')
plt.show()
```

```python
# 8. Distribucin de clase por puerto (Embarked)
plt.figure(figsize=(8, 5))
sns.countplot(x='Embarked', hue='Pclass', data=df)
plt.title('Distribucin de Clase por Puerto de Embarque')
plt.show()
```

```python
# 9. Conteo de pasajeros por gnero (Sex)
plt.figure(figsize=(6, 5))
sns.countplot(x='Sex', data=df)
plt.title('Conteo de Pasajeros por Gnero')
plt.show()
```

```python
# 10. Valores mximo y mnimo de edad
print(f"Edad Mnima: {df['Age'].min()}")
print(f"Edad Mxima: {df['Age'].max()}")
```

```python
# 11. Distribucin de edades con media, mediana y moda
mean_age = df['Age'].mean()
median_age = df['Age'].median()
mode_age = df['Age'].mode()[0]

plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], kde=True, color='lightgreen')
plt.axvline(mean_age, color='red', linestyle='--', label=f'Media: {mean_age:.1f}')
plt.axvline(median_age, color='green', linestyle='-', label=f'Mediana: {median_age:.1f}')
plt.axvline(mode_age, color='blue', linestyle='-.', label=f'Moda: {mode_age:.1f}')
plt.title('Distribucin de Edades con Estadsticas')
plt.legend()
plt.show()
```

```python
# 12. Diagrama de caja (boxplot) de edad (Outliers)
plt.figure(figsize=(8, 5))
sns.boxplot(x=df['Age'], color='orange')
plt.title('Boxplot de Edades (Detectar Outliers)')
plt.show()
```

```python
# 13. Valores mximo y mnimo de tarifa (Fare)
print(f"Tarifa Mnima: {df['Fare'].min()}")
print(f"Tarifa Mxima: {df['Fare'].max()}")
```

```python
# 14. Distribucin de tarifas con media, mediana y moda
mean_fare = df['Fare'].mean()
median_fare = df['Fare'].median()
mode_fare = df['Fare'].mode()[0]

plt.figure(figsize=(10, 6))
sns.histplot(df['Fare'], kde=True, color='salmon')
plt.axvline(mean_fare, color='red', linestyle='--', label=f'Media: {mean_fare:.1f}')
plt.axvline(median_fare, color='green', linestyle='-', label=f'Mediana: {median_fare:.1f}')
plt.axvline(mode_fare, color='blue', linestyle='-.', label=f'Moda: {mode_fare:.1f}')
plt.title('Distribucin de Tarifas con Estadsticas')
plt.legend()
plt.xlim(0, 300) # Limitar eje X para ver mejor la distribucin principal
plt.show()
```

```python
# 15. Diagrama de caja (boxplot) de tarifa (Fare)
plt.figure(figsize=(8, 5))
sns.boxplot(x=df['Fare'], color='cyan')
plt.title('Boxplot de Tarifas (Detectar Outliers)')
plt.show()
```

```python
# 16. Filtrar pasajeros con tarifa 0
pasajeros_gratis = df[df['Fare'] == 0]
print("\n--- Pasajeros con Tarifa 0 ---")
print(pasajeros_gratis[['Pclass', 'Sex', 'Age', 'Embarked']].head())
```

```python
# 17. Conteo de pasajeros por SibSp
plt.figure(figsize=(8, 5))
sns.countplot(x='SibSp', data=df)
plt.title('Conteo de Pasajeros por # de Hermanos/Esposos (SibSp)')
plt.show()
```

```python
# 18. Tasa de supervivencia basada en SibSp
plt.figure(figsize=(8, 5))
sns.barplot(x='SibSp', y='Survived', data=df, errorbar=None)
plt.title('Tasa de Supervivencia por SibSp')
plt.ylabel('Tasa de Supervivencia')
plt.show()
```

```python
# 19. Porcentaje de supervivencia para cada valor de SibSp
survival_sibsp = df.groupby('SibSp')['Survived'].mean() * 100
print("\n--- Porcentaje de Supervivencia por SibSp ---")
print(survival_sibsp)
```

```python
# 20. Mapa de calor (heatmap) de correlacin
plt.figure(figsize=(10, 8))

numeric_cols = df.select_dtypes(include=[np.number])
sns.heatmap(numeric_cols.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Matriz de Correlacin de Variables Numricas')
plt.show()
```
